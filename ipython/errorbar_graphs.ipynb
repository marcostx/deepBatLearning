# Estatísticas Gerais:  Accuracy, Precision, F1, Recall 

Neste notebook, são apresentados os gráficos de errorbar para as 4 métricas (accuracy, precision,f1 e recall). Os valores de cada gráfico representam a média obtida através de uma validação cruzada com k=10 folds, juntamente com o desvio padrão.

```python
import pandas as pd 
import numpy as np
import pandas
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
acc_cnn = [0.89610389610389607,0.86274509803921573,0.87581699346405228,
           0.88888888888888884,0.86184210526315785,0.89333333333333331,0.89932885906040272,
          0.891156462585034,0.89655172413793105,0.8896551724137931]
acc_mlp = [0.77272727272727271,0.82352941176470584,0.85620915032679734,
          0.82352941176470584,0.80921052631578949,0.79333333333333333,0.85906040268456374, 
           0.77551020408163263,0.7931034482758621,0.84827586206896555]
acc_dfa = [0.61879895561357701,0.63421052631578945,0.63324538258575203,
           0.60686015831134565,0.61741424802110823,0.60949868073878632,
           0.59366754617414252,0.6164021164021164,0.59946949602122013,0.6171276595744681]
plt.plot(acc_cnn,label="CNN")
plt.plot(acc_mlp,label='MLP')
plt.plot(acc_dfa,label='DFA')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# calculando as medias e desvios
mean_acc_ACU = np.mean(acc_cnn)
mean_mlp_ACC = np.mean(acc_mlp)
mean_dfa_ACC = np.mean(acc_dfa)
std_acc_ACC = np.std(acc_cnn)
std_mlp_ACC = np.std(acc_mlp)
std_dfa_ACC = np.std(acc_dfa)

mean_vals_acu = [mean_acc_ACU,mean_mlp_ACC,mean_dfa_ACC]
std_vals_acu = [std_acc_ACC,std_mlp_ACC,std_dfa_ACC]
print(mean_vals_acu)
```

```python
titles = ['CNN', 'MLP', 'DFA']
width=0.35
y_pos  = np.arange(len(titles))

fig, ax = plt.subplots()
fig.set_size_inches(15.5, 10.5)
ax.bar(y_pos, mean_vals_acu, width,yerr=std_vals_acu, align='edge',
        color='red', ecolor='black')
ax.set_xticks(y_pos + width/2)
ax.set_xticklabels(('CNN', 'MLP', 'DFA'))
plt.ylim(0.0, 1.0)
ax.set_xlabel('Model')
ax.set_ylabel('Mean')
ax.set_title('Accuracy')
plt.savefig('accuracy.png')
plt.show()
```
